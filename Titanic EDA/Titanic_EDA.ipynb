```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing and Configuring DataSet

```python
train = pd.read_csv("train_and_test2.csv")
```

```python
train.head(10)
```

```python
train.tail(10)
```

```python
train.describe()
```

```python
train['Gender'] = train['Sex'].apply(lambda x: 'M' if x==1 else 'F')
```

```python
del train['Sex']
```

```python
train['survived'] = train['2urvived'].apply(lambda x: 'Y' if x==1 else 'N')
```

```python
del train['2urvived']
```

```python
columns = list(train.columns.values)
columns
```

```python
train = train[['Passengerid','Age','Fare', 'Gender', 'survived', 'sibsp', 'zero', 'zero.1','zero.2', 'zero.3', 'zero.4', 'zero.5', 'zero.6','Parch', 'zero.7', 'zero.8', 'zero.9',
 'zero.10','zero.11', 'zero.12', 'zero.13','zero.14', 'Pclass', 'zero.15', 'zero.16','Embarked', 'zero.17', 'zero.18']]
```

```python
train.drop(columns = ['zero', 'zero.1','zero.2', 'zero.3', 'zero.4', 'zero.5', 'zero.6', 'zero.7', 'zero.8', 'zero.9',
 'zero.10','zero.11', 'zero.12', 'zero.13','zero.14', 'zero.15', 'zero.16', 'zero.17', 'zero.18'], inplace = True)
```

```python
train.head()
```

# Checking and Removing Null Values

```python
train.isnull().sum()
```

```python
sns.heatmap(train.isnull())
```

```python
train.dropna(inplace = True) #dropping because null values are neglibile and won't hurt accuracy
```

# Analysis begins

#### Age Analysis

```python
train.Age.unique()
```

```python
sns.kdeplot(train['Age'])
```

```python
#Age distribution in the passengers
sns.histplot(x = 'Age', data = train, kde = True, hue = 'Gender', palette = "YlGnBu", bins = 20)
```

```python
train[['Age', 'Gender']]
```

```python
#Age distribution according to Gender
sns.catplot(y = 'Age', data = train, hue = 'Gender', dodge = True)
```

```python
sns.catplot(x = 'Age', y = 'Fare', data = train, hue = 'survived')
```

#### Survivor Analysis 

```python
#Tells us about the number of survivors
sns.countplot(x = 'survived', data = train, palette = 'YlGnBu')
```

```python
#Tells us about the number of survivors according to gender
sns.countplot(x = 'survived', data = train, hue = 'Gender', palette = 'YlGnBu')
```

```python
#Shows the relation between the number of survivors and which class they were staying in
sns.countplot(x='survived',hue='Pclass',data=train)
```

```python
train2 = train.copy()
train2.set_index('survived', inplace = True)
```

```python
train2.loc['Y']
```

```python
#Survivors according to age
sns.stripplot(x = 'survived', y = 'Age', data = train2, dodge = True, hue = 'Gender', palette = 'YlGnBu')
```

#### Embarkment Analysis

```python
#Total number of people and their port of Embarkment
sns.countplot(x = 'Embarked', data = train)
```

```python
#Tells us about the gender ratio at every port of embarkment
sns.countplot(x = 'Embarked', hue = 'Gender', palette = 'YlGnBu', data = train)
```

```python
#Shows how many people from which location survived/didn't survive
sns.countplot(x = 'Embarked', hue = 'survived', data = train, palette = 'YlGnBu')
```
